```python
from pathlib import Path
import elevenlabs
import IPython
```

```python
my_cwd = Path().resolve()
```

```python
elevenlabs_key = my_cwd / 'elevenlabs'
f = open(elevenlabs_key, "r")
api_key = f.read()
elevenlabs.set_api_key(api_key)
```
